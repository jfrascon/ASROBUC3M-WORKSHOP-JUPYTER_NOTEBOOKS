```python
%matplotlib inline
import matplotlib.pyplot as plt

# +---------------------------------------------+
# | LINEAR REGRESSION AND POLYNOMIAL REGRESSION |
# +---------------------------------------------+
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0.1, 0.2, 0.3, 0.5, 0.8, 2.0])

p1 = np.polyfit(x, y, 1) # linear
p2 = np.polyfit(x, y, 2) # quadratic
p3 = np.polyfit(x, y, 3) # cubic

print(p1)
print(p2)
print(p3)

plt.plot(x, y, 'o')
plt.plot()

xp = np.linspace(0, 5, 100)
plt.plot(xp, np.polyval(p1, xp), 'r-', label='linear')
plt.plot(xp, np.polyval(p2, xp), 'g-', label='quadratic')
plt.plot(xp, np.polyval(p3, xp), 'k-', label='cubic')


# +----------------------+
# | NONLINEAR REGRESSION |
# +----------------------+
from scipy.optimize import curve_fit

def f(x, a):
    return 0.1 * np.exp(a*x)

# The first value is the value of the 'a' parameter
# The second value is the covariance of the error.
a,error_cov = curve_fit(f, x, y)
plt.plot(xp, f(xp, a), 'c-', label='non-linear')
plt.legend(loc='best')
```

```python
# +-----------------------+
# | DIFFERENTIAL EQUATION |
# +-----------------------+

from scipy.integrate import odeint

# dy/dt = -0.4 * y
def dydt(y, t, a):
    return a*y

a = -0.4
t = np.linspace(0, 20)
y0 = 5.0 # initial condition
y = odeint(dydt, y0, t, args=(a,))

plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y')
plt.xlim([0, 20])
plt.ylim([0, 5])
```
